```python
# 先取5w个试一试 :去掉字母
length=200000
def read_data():
    data = {}
    x, y = [], []
  
    
    for i in range(length):
        temp=text.loc[i].train.split()
        index=[]
        for i in range(len(temp)):
            if zhmodel.search(temp[i]):
                index.append(i)
        temp=[temp[i] for i in index]
        x.append(temp)
        y.append(text.loc[i].label)
    x, y = shuffle(x, y)
    dev_idx =int(length* 0.8    )
    test_idx =int( length* 0.9  )

    data["train_x"], data["train_y"] = x[:dev_idx], y[:dev_idx]
    data["dev_x"], data["dev_y"] = x[dev_idx:test_idx], y[dev_idx:test_idx]
    data["test_x"], data["test_y"] = x[test_idx:], y[test_idx:]


    return data
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.BATCH_SIZE = batch_size
        self.MAX_SENT_LEN = max_seq_len
        self.WORD_DIM = WORD_DIM
        self.VOCAB_SIZE = VOCAB_SIZE
        self.CLASS_SIZE = CLASS_SIZE
        self.DROPOUT_PROB = drop_out
        self.hidden_dim=80

        # one for UNK and one for zero padding
        self.embedding = nn.Embedding(self.VOCAB_SIZE + 2, self.WORD_DIM, padding_idx=self.VOCAB_SIZE + 1)
        self.embedding.weight.data.copy_(torch.from_numpy(wv_matrix)) 
        self.embedding.weight.requires_grad = False  # static 
        
        self.bigru = nn.GRU( self.WORD_DIM, self.hidden_dim, dropout= DROPOUT_PROB, num_layers=1, bidirectional=True)
        
        self.hidden2label = nn.Linear(self.hidden_dim * 4,  self.CLASS_SIZE)
        

    def forward(self, inp):
        x = embedding(inp)
        x = x.view(x.size(1),len(inp) , -1) 
        x = F.dropout(x, p=self.DROPOUT_PROB, training=self.training)
        gru_out, _ = bigru(x) 
        gru_out = torch.transpose(gru_out, 1, 2) 
        gru_out = torch.transpose(gru_out, 0, 2)   #torch.Size([32, 160, 30])
        max_out = F.max_pool1d(gru_out, gru_out.size(2)).squeeze(2) 
        average_out = F.avg_pool1d(gru_out, gru_out.size(2)).squeeze(2) 
        conc =torch.cat((max_out,average_out),1)

        
        y = self.hidden2label(conc)
        logit = y
        
        return logit
```
